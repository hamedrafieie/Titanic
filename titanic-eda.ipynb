```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='darkgrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.displot(titanic, x='age')
```

```python
sns.displot(titanic, x='age', hue='sex',multiple="dodge")
```

```python
sns.catplot(data=titanic, x="sex", y="survived", hue="class", kind="bar")
```

```python
sns.catplot(data=titanic, x="survived", y="deck", errorbar=("pi", 95), kind="bar")
```

```python
g = sns.catplot(
    data=titanic,
    x="fare", y="embark_town", row="class",
    kind="box", orient="h",
    sharex=False, margin_titles=True,
    height=1.5, aspect=4,
)
g.set(xlabel="Fare", ylabel="")
g.set_titles(row_template="{row_name} class")
for ax in g.axes.flat:
    ax.xaxis.set_major_formatter('${x:.0f}')
```

```python
sns.catplot(data=titanic, x="pclass", y="survived", errorbar=("pi", 95), kind="bar")
```

```python
sns.boxplot(x='survived', y='fare', data=titanic, showfliers=False)
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.title('Fare Distribution by Survival')
plt.show()
```

```python
sns.histplot(data=titanic, x='fare', hue='survived', kde=True, bins=20)
plt.xlabel('Fare')
plt.ylabel('Count')
plt.title('Fare Distribution by Survival')
plt.show()
```
